```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
telco_data=pd.read_csv('Churn_Modelling.csv')
```

```python
telco_data.head()
```

```python
telco_new=telco_data[['Geography','Gender','Exited']]
```

```python
telco_new
```

### Univariate Analysis

```python
for i,predictor in enumerate(telco_new.drop(columns=['Exited'])):
    plt.figure()
    sns.countplot(data=telco_new,x=predictor,hue='Exited')
```

```python
telco_new.Exited.value_counts()
```

```python
telco_new.Exited.value_counts()/len(telco_new)*100
```

### Bivariate Analysis

```python
sns.histplot(x='Gender',hue='Geography',data=telco_new,stat='count',multiple='dodge')
```

```python
telco_new_target1=telco_new.loc[telco_new['Exited']==1]
```

```python
sns.histplot(x='Gender',hue='Geography',data=telco_new_target1,stat='count',multiple='dodge')
```

### Numerical Analysis

##### Correlation:It ranges from -1 to +1

```python
plt.figure(figsize=(20,8))
telco_data.corr(numeric_only=True)['Exited'].sort_values(ascending=False).plot(kind='bar')
```

```python
telco_data.corr(numeric_only=True)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(telco_data.corr(numeric_only=True),cmap='Paired')
```

```python
Tot=sns.kdeplot(telco_data.Age[(telco_data['Exited']==0)],color='green',shade=True)
Tot=sns.kdeplot(telco_data.Age[(telco_data['Exited']==1)],color='red',shade=True)
Tot.legend(['No Churn','Churn'],loc='upper right')
Tot.set_xlabel('Age')
Tot.set_ylabel('Density')
Tot.set_title('Age by churn')
```

```python
Tot=sns.kdeplot(telco_data.CreditScore[(telco_data['Exited']==0)],color='green',shade=True)
Tot=sns.kdeplot(telco_data.CreditScore[(telco_data['Exited']==1)],color='red',shade=True)
Tot.legend(['No Churn','Churn'],loc='upper right')
Tot.set_xlabel('Credit Score')
Tot.set_ylabel('Density')
Tot.set_title('Credit Score by churn')
```

```python
Tot=sns.kdeplot(telco_data.Balance[(telco_data['Exited']==0)],color='green',shade=True)
Tot=sns.kdeplot(telco_data.Balance[(telco_data['Exited']==1)],color='red',shade=True)
Tot.legend(['No Churn','Churn'],loc='upper right')
Tot.set_xlabel('Balance')
Tot.set_ylabel('Density')
Tot.set_title('Balance by churn')
```
