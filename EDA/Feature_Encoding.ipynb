### Feature Encoding

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.info()
```

```python
## Handle missing values if missing values are present
```

#### Label Encoding

```python
le=preprocessing.LabelEncoder()
df['Gender_label']=le.fit_transform(df.Gender.values)
```

```python
df.head(5)
```

```python
df.Gender_label.value_counts()
```

#### One Hot Encoding

```python
ohe=preprocessing.OneHotEncoder(sparse_output=False)
encoded=ohe.fit_transform(df[['Geography']])
```

```python
encoded
```

```python
df_encoded=pd.DataFrame(encoded,columns=ohe.get_feature_names_out())
```

```python
df_encoded
```

```python
## Alternate and easy way
one_hot=pd.get_dummies(df['Geography'])
```

```python
one_hot
```

```python
## need to perform encoding all dataframe
df_dummies=pd.get_dummies(df)
df_dummies
```

#### Dummy Encoding
It will delete the dummy column that is if 'n' dummy columns are created through this we can delete any one of the dummy column and make it as '(n-1)' dummy columns

```python
df_dummies_de=pd.get_dummies(df,drop_first=True) ## or drop='first'
df_dummies_de.head(5)
```
