## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
## datapoints are linearly seperable
# X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0,random_state=15)
```

```python
## if we have by default n_clusters_per_class=2 then the data points are not linearly seperable
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=15)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
##Prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
svc.coef_
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
##Prediction
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
## poly is giving bad accuracy than linear for the above dataset
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
```

```python
## Prediction
y_pred2=polynomial.predict(X_test)
```

```python
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
## sigmoid also giving bad accuracy for the above dataset
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
```

```python
y_pred3=sigmoid.predict(X_test)
```

```python
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_,
```

### Conclusion: 'rbf' Kernel is best for the above kind of dataset (where n_clusters_per_class value is 2) when compared to 'polynomial' and 'sigmoid'  

# HYPERPARAMETER TUNING With SVC

```python


from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
'gamma':[1,0.1,0.001,0.0001],
'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
