# Naive Baye's Algorithm Implementation

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
## Independent and dependent features
X=df[['tip', 'sex', 'total_bill', 'day', 'time', 'size']]
y=df['smoker']
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train
```

```python
## One Hot Encoding---- ColumnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols=[ 'sex', 'day', 'time']
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),categorical_cols)],remainder='passthrough')
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```
