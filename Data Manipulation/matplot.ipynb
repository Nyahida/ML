### Matplotlib

```python
import matplotlib.pyplot as plt
```

#### Line Plot

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot
plt.plot(x,y,color='green',linestyle=':',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## multiple plots

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='red')
plt.title("plot 1")
plt.subplot(2,2,2)
plt.plot(y1,y2,color='blue')
plt.title("plot 2")
plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("plot 3")
plt.subplot(2,2,4)
plt.plot(x,y2,color='red')
plt.title("plot 4")
```

#### Bar Plot

```python
categories=['A','B','C','D']
values=[10,13,8,20]

plt.bar(categories,values,color='purple')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Bar Plot")
plt.show()
```

#### Histogram

```python
data=[1,1,2,3,2,4,4,8,8,8,8,4,6,3,5,5,8,7,6]

plt.hist(data,bins=5,color='yellow',edgecolor='black')
```

```python
x=[1,2,3,4,5]
y=[1,4,2,6,5]

plt.scatter(x,y,color='green',marker='*')
```

#### Pie Chart

```python
labels=['A','B','C','D']
sizes=[10,13,8,20]
colors=['gold','red','blue','green']

explode=(0.2,0,0,0)

plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,explode=explode)
```

```python
import pandas as pd
```

```python
crkt_csv=pd.read_csv('cricket.csv')
crkt_csv
```

```python
crkt_csv_playing_role=crkt_csv.groupby('PLAYING ROLE')['SOLD PRICE'].sum()
print(crkt_csv_playing_role)
```

```python
crkt_csv_playing_role.plot(kind='bar',color='teal')
```

```python
crkt_auct_yr=crkt_csv.groupby('AUCTION YEAR')['SIXERS'].sum().reset_index()
print(crkt_auct_yr)
```

```python
plt.plot(crkt_auct_yr['AUCTION YEAR'],crkt_auct_yr['SIXERS'])
```
