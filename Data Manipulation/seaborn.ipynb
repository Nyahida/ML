### Seaborn


```python
import seaborn as sns
```

```python
## Basic Plotting with Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
### create scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot")
plt.show()
```

```python
## Categorial Plots
sns.barplot(x='day',y='total_bill',data=tips,err_kws={'color':'g','linewidth':5},color='red')
plt.title("Bar Plot")
plt.show()
```

```python
## Box Plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot")
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## Histogram
sns.histplot(tips['total_bill'],bins=8,kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
### Pair Plot
sns.pairplot(tips,diag_kind='kde')
```

```python
### Pair Plot
sns.pairplot(tips,diag_kind='kde',x_vars=['total_bill','size'],y_vars=['tip','size'],kind='reg')
```

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
